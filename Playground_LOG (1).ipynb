```python
attrition = pd.read_csv(io.BytesIO(uploaded['Attrition.csv']))
attrition.head(3)
```

```python
# rename column name
attrition.rename(columns = {'EmployeeNumber':'id'}, inplace = True)
attrition.head(3)
```

```python
# replace yes/no
attrition.loc[attrition['Attrition'] == 'Yes','Attrition'] = 1
attrition.loc[attrition['Attrition'] == 'No','Attrition'] = 0
attrition['Attrition'].unique()
```

```python
# concat data
attrition = attrition[train.columns.tolist()]
frames = [train, attrition]
train = pd.concat(frames)
train.reset_index( drop=True, inplace=True)
train
```

```python
features =  list(train.columns)
features.remove('id')
features.remove('Attrition')

target = 'Attrition'
```

```python
columns_with_strings_as_values = list((train.dtypes[train.dtypes == 'object']).index)
columns_with_strings_as_values
```

```python
columns_with_strings_as_values =['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime']
columns_with_strings_as_values
```

```python
from sklearn.preprocessing import LabelEncoder
for col in columns_with_strings_as_values:
  le = LabelEncoder()
  train[col] = le.fit_transform(train[col])
  test[col] = le.fit_transform(test[col])

train.head(3)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
# Get the split indexes
strat_shuf_split = StratifiedShuffleSplit(n_splits=1,
                                          test_size=0.3,
                                          random_state=42)

train_idx, test_idx = next(strat_shuf_split.split(train[features],train.Attrition))

#create the dataframe
X_train = train.loc[train_idx, features]
y_train = train.loc[train_idx, 'Attrition']
y_train=y_train.astype('int')

X_test = train.loc[test_idx,features]
y_test = train.loc[test_idx,'Attrition']
y_test=y_test.astype('int')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
# standard logistic regression
lr_l1 = LogisticRegressionCV(Cs=10,cv=4,penalty='l1',solver='liblinear').fit(X_train, y_train)
```

```python
# predict the class and the probabiity for each
y_pred = list()
y_prob = list()
y_pred.append(pd.Series(lr_l1.predict(X_test)))
y_prob.append(pd.DataFrame(lr_l1.predict_proba(X_test)))


y_pred = pd.concat(y_pred, axis=1)
y_prob = pd.concat(y_prob, axis=1)

y_pred
```

```python
y_prob
```

```python
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.preprocessing import label_binarize

metrics = list()

precision, recall, fscore, _ = score(y_test,y_pred, average='weighted')
accuracy = accuracy_score(y_test, y_pred)
metrics.append(pd.Series({'precision':precision,'recall':recall, 'fscore':fscore, 'accuracy':accuracy}))
  
metrics = pd.concat(metrics, axis=1)
metrics
```
